# ― 線形代数 ― Linear Algebra

## スカラー・ベクトル・行列について

### 【スカラー】scalar

$x, y, z, m, n $

-  サイズ：1✕1のスカラー

### 【ベクトル】vector

$ \mathbf{x} = \left(\begin{array}{c} x_1 \\ x_2 \\ x_n \end{array}\right)
,\ \mathbf{y} = \left(\begin{array}{c} y_1 \\ y_2 \\ \vdots \\ y_n \end{array}\right) $

または

$ \mathbf{x} =　\begin{bmatrix}
x_1 \\
x_2 \\
x_3 \\
\end{bmatrix}
,\ \mathbf{y} =　\begin{bmatrix}
y_1 \\
y_2 \\
\vdots \\
y_n \\
\end{bmatrix}
$

- 丸括弧、角括弧どちらでもOK

- スカラーが複数個集まったものをベクトルと呼ぶ(データを塊として扱える)

- $\mathbf{x}$ これはxの太文字を表す

- イメージとしてはエクセルから1カラム取ってきたようなもの

- xサイズ：3✕1のベクトル

- yサイズ：N✕1のベクトル

- 行列の一部としてベクトル、ベクトルの一部としてスカラーがある

#### ベクトルの転置について

$ \mathbf{x}^T = [1, 2, \ldots, n] $

- 横向きのベクトルの場合は T(転置) のマークを付ける

### 【行列】matrix

$ \mathbf{X} =　\begin{bmatrix}
x_{11} & x_{12} \\
x_{21} & x_{22} \\
x_{31} & x_{32} \\
\end{bmatrix}
$

- ベクトルに対して更に横方向を持ったものを行列と呼ぶ

- サイズ：3行2列（例）　3✕2の行列と呼ぶ

- $\mathbf{X}$ これはXの太文字を表す

##### ※ 行列をさらに並べたものを"テンソル"(tensor)と呼ぶ

##### ※文字サイズについて

- 細文字　＋　小文字　スカラーの変数　$ x $

- 細文字　＋　大文字　スカラーの定数 $ X $

- 太文字　＋　小文字　ベクトル $ \mathbf{x} $

- 太文字　＋　大文字　行列 $ \mathbf{X} $

### 足し算、引き算

- (ポイント）サイズが同じであることが条件

#### ※ 行列A (N行,M列)  ✕  行列B (0行,P列) = 行列C (N行,P列)

- (条件) M = O であること

## 転置 Transpose

$ \mathbf{x} = \begin{bmatrix}1 \\ 2 \\ 3 \\ \end{bmatrix} $

$ \mathbf{x}^T = \begin{bmatrix}1 & 2 & 3 \end{bmatrix} $

$ \mathbf{X} = \begin{bmatrix}
1 & 4 \\
2 & 5 \\
3 & 6 \\
\end{bmatrix} $

$ \mathbf{X}^T = \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6 \\
\end{bmatrix} $

### 転置の公式

- (1) $ (A^T)^T = A $


- (2) $ (AB)^T = B^TA^T $


- (3) $ (ABC)^T = C^T B^T A^T $

## 単位行列と逆行列

### 【単位行列】 Unit matrix

スカラーの[1]に対応

2 * 1 = 2 →　1を掛ける前と一緒

#### A * I = A

#### I * A = A

$ \mathbf{I} = \begin{bmatrix}
1 & 0 & 0 & \ldots & 0 \\
0 & 1 & 0 & \ldots & 0 \\
\vdots & \vdots & \vdots & \ddots & \vdots \\
0 & 0 & 0 & \ldots & 1 \\
\end{bmatrix} $

$ \mathbf{I}_{2✕2} = \begin{bmatrix}
1 & 0 \\
0 & 1 \\
\end{bmatrix} $

$ \mathbf{I}_{3✕3} = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1 \\
\end{bmatrix} $

### 【逆行列】 Inverse matrix

スカラーの逆数に対応するものを逆行列と呼ぶ

(例)

$ 2 * ( \ ) = 1 $

$ 2 * (1/2) = 1 $

$ 2 * (2^{-1}) = 1 $

よって

$ \mathbf{A} * ( \ ) = \mathbf{I} $ 

$ \mathbf{A} * (\mathbf{A}^{-1}) = \mathbf{I} $

$ (\mathbf{A}^{-1}) * \mathbf{A} = \mathbf{I} $

- $ \mathbf{I} $ = 単位行列

- $ \mathbf{A}^{-1} $= Aの逆行列 (Aインバースと呼ぶ)

#### (条件)

- 行列Aが "正方行列" であること (最低限)

## ベクトルで微分する

$ \mathbf{c} = 
\begin{bmatrix}
3 \\
4 \\
\end{bmatrix}
,\ \mathbf{x} = 
\begin{bmatrix}
x_1 \\
x_2 \\
\end{bmatrix}
$ のとき、

$ \mathbf{c}^T \mathbf{x} = 
\begin{bmatrix}
3 & 4 \\
\end{bmatrix}
\begin{bmatrix}
x_1 \\
x_2 \\
\end{bmatrix}
= 3x_1 + 4x_2 $

- $\mathbf{c}^T \mathbf{x} $ ←スカラー

$ \displaystyle \frac{\partial}{\partial x}(\mathbf{c}^T \mathbf{x}) =
\begin{bmatrix}
\displaystyle \frac{\partial}{\partial x_1}(3x_1 + 4x_2) \\
\displaystyle \frac{\partial}{\partial x_2}(3x_1 + 4x_2) \\
\end{bmatrix}$

##### (1) まず$x_1$を微分する

$ \displaystyle \frac{\partial}{\partial x_1}(3x_1 + 4x_2) =
\displaystyle \frac{\partial}{\partial x_1}(3x_1) +
\displaystyle \frac{\partial}{\partial x_1}(4x_2)$

$ = 3 × \displaystyle \frac{\partial}{\partial x_1}(x_1) +
4x_2 × \displaystyle \frac{\partial}{\partial x_1}(1)$

$ = 3 $

##### 微分の公式を適用

- $ \displaystyle \frac{\partial}{\partial x_1}(x_1) → 1 $

- $ \displaystyle \frac{\partial}{\partial x_1}(1) → 0 $

##### (2) 次に$x_2$を微分する

$ \displaystyle \frac{\partial}{\partial x_2}(3x_1 + 4x_2) =
\displaystyle \frac{\partial}{\partial x_1}(3x_1) +
\displaystyle \frac{\partial}{\partial x_1}(4x_2)$

$ = 3x_1 × \displaystyle \frac{\partial}{\partial x_2}(1) +
4 × \displaystyle \frac{\partial}{\partial x_2}(x_2)$

$ = 4 $

##### (3) 答えは...

$ \mathbf{c}^T \mathbf{x} =
\begin{bmatrix}
3 \\
4 \\
\end{bmatrix} $

最初の $\mathbf{c} = 
\begin{bmatrix}
3 \\
4 \\
\end{bmatrix}$ と同じ!!!! 転置が取れて元に戻ってる！

#### 例題

$ \displaystyle \frac{\partial}{\partial \mathbf{x}}(c) =
\begin{bmatrix}
\displaystyle \frac{\partial}{\partial x_1}(c) \\
\displaystyle \frac{\partial}{\partial x_2}(c) \\
\end{bmatrix} =
\begin{bmatrix}
0 \\
0 \\
\end{bmatrix}$

- c = 定数　(１， 2, … )

$\begin{bmatrix}
0 \\
0 \\
\end{bmatrix} = \mathbf{0} $ ←　これを0ベクトルと呼ぶ



## ベクトルで微分の公式

### (1) $ \displaystyle \frac{\partial}{\partial \mathbf{x}}(c) = \mathbf{0} $

- c = 定数

### (2) $ \displaystyle \frac{\partial}{\partial \mathbf{x}}(\mathbf{b}^T \mathbf{x}) = \mathbf{b} $

### (3) $ \displaystyle \frac{\partial}{\partial \mathbf{x}}(\mathbf{x}^T \mathbf{A} \mathbf{x}) =
(\mathbf{A} + \mathbf{A}^T) \mathbf{x} $
