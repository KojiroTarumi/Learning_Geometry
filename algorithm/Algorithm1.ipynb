演習課題
素数を見つけるアルゴリズム

- 素数を格納するnumbersという空のリストで初期化
- iを1〜10000まで繰り返す

 - 素数判定フラグをTrueに初期化
 - jを 2,..., i-1　の数値まで割り切れたら、フラグを False + その時点で抜ける(break)
 - iより小さなすべての値(1は除く)で計算して、フラグが Trueのママであれば素数
 - 素数であればnumbersに追加
 
例えば10のとき

- 2, 3, 4, 5, 6, 7, 8, 9 で割り算してあまりをチェック
- もし1つでも余りが0となっていたら割り切れているので素数ではない
- すべてのあまりが0でなければ素数

```python
#10,000までの素数を求める
#アルゴリズム
#numberをすべての整数で割る(試し割り法)
#コード→リファクタリング

def get_prime_numbers(max_val):
    numbers = []

    for i in range(1, max_val):
        flag = True
        for j in range(2, i):
            if i % j == 0:
                flag = False
                break
            
        if flag == True:
            numbers.append(i)

    return numbers

numbers = get_prime_numbers(10000)
print(numbers)
```

```python
#リストに対する和を計算する関数'add'を作る
a = [1, 2, 3]
b = [4, 5, 6]

#add(a, b) → [5, 7, 9]
# -まず動かす
# -リファクタリングする

#要素ごとに取り出して足す

def add(a, b):
    c = [a[0] + b[0], a[1] + b[1], a[2] + b[2]]
    return c

ngo = add(a, b)
print(ngo)
```

```python
#汎用性を高める

#空リスト定義
# i = 0, 1, 2
#  -- a[i], b[i]の足し算を行う
#  -- resultsに結果を追加

a = [1, 2, 3]
b = [4, 5, 6]

def add(a, b):
    results = []
    for i in range(len(a)):    #←3だと3個にしか対応していないのでもっと可読性を高める必要がある len(a)
        results.append(a[i] + b[i])
        # result = a[i] + b[i]
        #results.append(result) ←この程度ならappendに直接入れてもok
    return results

add(a, b)
```

```python
#可読性を高める

def add(a, b):
    results = []
    for (_a, _b) in zip(a, b):    #zip関数はaから一つ取り出して_a(x)に入れる、bから一つ取り出して_b(y)に入れる事ができる
        results.append(_a + _b)　　　　#要素番号抜きで使える
#    for i in range(len(a)):
#        results.append(a[i] + b[i])
    return results

add(a, b)
```

```python
#例外処理
#add(c, d)の場合errorを表示する
#先程の関数を利用してその外側にif文で分岐を作る
#if文で要素数が等しい場合と等しくない(error)の場合に分岐させる

c = [1, 2, 3]
d = [4, 5, 6, 7]

def add(c, d):
    if len(c) == len(d):
        results = []
        for (_c, _d) in zip(c, d): 
            results.append(_c + _d)
        return results
    else:
        return 'error'
    
add(c, d)
```

**問題**：オリジナルファイルtextを情報が劣化しないように圧縮する
#1) 圧縮を行う関数'encode'を作る　text_encode
#2) 復元を行う関数'decode'を作る text_decode == text

まずアルゴリズム

  圧縮の仕組み：aaaaabbbbbccc →　a5b5c3
  




```python
with open('input.txt', 'r')as f:
    text = f.read()

row = [text[:1000]]

#for文でtextの1〜1000までの数を出力する  ←NG!!!!!!!!
    #if文で次の数が0ではない場合一旦停止
        #0の後にそこまでの数値を記載する
        #continueで1から0まで繰り返す
```

解答

text = '00000011111111'

- 000000 →0[06]
- 111111111111 →1[12]

text_encode = '006112'

0 or 1 の2択

text_encode = '00612' ←0の次は必ず1なのでもっと短くできる

(encode)
text_encode = encode(text)
text_encode → '00612'

(decode)
text_decode = decode(text_encode)
text_decode → '000000111111111111'

アルゴリズム
1 先頭が0か1を取得
2 何文字続くかカウントを取得（○○文字)
3 一桁の場合は2行に0を入れて補正
4 整数を文字列にする
5 すべてを文字として足し合わせる

```python
def get_count(text):
    #先頭
    _head = text[0]    #テンポラリに一次避難することでなにか数値が変わっても処理することが可能
    head = _head

    counts = []
    count = 0
    for ch in text:    #chはキャラクター
        if ch == head:
            count += 1
        else:
            counts.append(count)
            count = 1
            head = ch #この時点でheadは1
    #for文は最後にelseで片付けができる
    else:
        counts.append(count)

    return _head, counts

head, counts = get_count(text)
head
```

```python
def int2str(count):    #関数化する　　　　2はtoの省略
    # 整数(6) → 文字列('06')に変換
    if count < 10:
        return '0' + str(count)
    else:
        return str(count)

int2str(6)
```

```python
#すべてつなげる
text_encode = head
for count in counts:
    text_encode += int2str(count)

text_encode
```

```python
#最後に関数化する
def encode(text):
    head, counts = get_count(text)

    text_encode = head
    for count in counts:
        text_encode += int2str(count)

    return text_encode

text
```

decodeのアルゴリズム

1 先頭(head)と本文(body)に分ける
2 bodyを2文字ずつに分割　→'0612' '06', '12'
3 文字列を整数に変換　→ '06' '6'
4 headの文字から交互にカウント分だけ文字列を足し合わせる

```python
text_encode = '00612'

head = text_encode[0]
body = text_encode[1:]

for i in range(0, len(body), 2):
    num = int(body[i: i+2])
    print(num)
```

```python
text_decode = ''

for i in range(0, len(body), 2):
    num = int(body[i: i+2])
    text_decode += head * num
    #文字列を交互に
    if head == '0':
        head = '1'
    else:
        head = '0'

text_decode
```

```python
def decode(text_encode):

    head = text_encode[0]
    body = text_encode[1:]
    
    text_decode = ''

    for i in range(0, len(body), 2):
        num = int(body[i: i+2])
        text_decode += head * num
        #文字列を交互に
        if head == '0':
            head = '1'
        else:
            head = '0'
    return text_decode
```
